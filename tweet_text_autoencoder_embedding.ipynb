```python
import keras

model_path = "Textual features\\Tweets\\autoencoder.h5"
autoencoder = keras.models.load_model(model_path)
encoder = keras.Model(autoencoder.input, autoencoder.layers[-5].output)

print(encoder.summary())
```

```python
import numpy as np
import statistics
from tqdm import tqdm

def tweet_vectorization(data):

  embed_vectors = {}

  print("Tweet Vectorization")
  for item in tqdm(data.items()):
    prof_name = item[0].replace(".json", "")
    tweets_vec = item[1]

    tweet_text_vec = encoder.predict(np.array(tweets_vec).reshape(-1,384))
    tweet_property_vec = [statistics.mean(k) for k in zip(*tweet_text_vec)]
    temp = list(tweet_property_vec)
    embed_vectors[prof_name] = [np.float64(x) for x in temp]

  return embed_vectors
```

```python
import json

def data_vectorization(tweets):

  tweet = tweet_vectorization(tweets)
  
  json_obj = json.dumps(tweet, indent=4)
  with open("tweets_embeddings_compressed.json", 'w', encoding="utf-8") as f:
    f.write(json_obj)
  f.close()

  return tweet
```

```python
import pickle

path = "Textual features\\Tweets\\tweets"

with open(path, 'rb') as fp:
  X = pickle.load(fp)

```

```python
output = data_vectorization(X)
```
