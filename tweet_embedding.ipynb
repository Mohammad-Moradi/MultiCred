```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import datetime
import numpy as np
import re
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler

# Function to clean a text from hashtags, @s and urls
def clean(text):
  clean_tweet = re.sub("@[A-Za-z0-9_]+","", text)
  clean_tweet = re.sub("#[A-Za-z0-9_]+","", clean_tweet)
  clean_tweet = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', clean_tweet)
  return clean_tweet


# Function to convert string to datetime
def convert(date_time):
    date_time = date_time.replace('+0000', "")
    format = '%a %b %d %H:%M:%S %Y'
    datetime_str = datetime.datetime.strptime(date_time, format)

    return datetime_str
```

```python
def tweet_vectorization(data):

  dic = {}
  ls = []

  print("Tweet Vectorization")
  for item in tqdm(data.items()):
    prof_name = item[0].replace(".json", "")
    tweets = item[1]

    aggregated_vector = []

    for tweet in tweets:

      text = clean(tweet["text"])
      text = text.lower()
      # tweet_text_vec = model.encode(text)
      aggregated_vector.append(text)

    tweet_text_vec = model.encode(aggregated_vector)
    ls.append(tweet_text_vec)
    dic[prof_name] = tweet_text_vec
  return dic
```

```python
import pickle

def data_vectorization(tweets):

  tweet = tweet_vectorization(tweets)

  with open('Textual features\\Tweets\\tweets_embeddings', 'wb') as fp:
    pickle.dump(tweet, fp)

  print('Done writing list into a binary file')

  return tweet
```
